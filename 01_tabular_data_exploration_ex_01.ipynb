How many features are numerical? How many features are categorical?

```python
num_features = penguins.select_dtypes(include=['float64', 'int64']).columns.tolist()
cat_features = penguins.select_dtypes(include=['object']).columns.tolist()
print("Numerical Features:", num_features)
print("Categorical Features:", cat_features)
```
